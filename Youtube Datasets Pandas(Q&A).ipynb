## 4. Views Analysis

```python
# Highest-viewed video
max_views = yt['views'].max()
yt[yt['views'] == max_views]
```

```python
# Least-viewed video
min_views = yt['views'].min()
yt[yt['views'] == min_views]
```

```python
# Top 5 most-viewed videos
yt.nlargest(5, 'views')
```

```python
# Top 5 least-viewed videos
yt.nsmallest(5, 'views')
```

```python
# Average views
yt['views'].mean()
```

```python
# Median views
yt['views'].median()
```

```python
# Total views across all videos
yt['views'].sum()
```

## 5. Likes Analysis

```python
# Most-liked video
max_likes = yt['likes'].max()
yt[yt['likes'] == max_likes]
```

```python
# Top 5 most-liked videos
yt.nlargest(5, 'likes')
```

```python
# Average likes
yt['likes'].mean()
```

```python
# Correlation between views and likes
yt[['views', 'likes']].corr()
```

## 6. Channel Analysis

```python
# Channel with the most uploads
yt['Channel_title'].value_counts().idxmax()
```

```python
# Top 10 channels by number of videos uploaded
yt['Channel_title'].value_counts().head(10)
```

```python
# Channel with the highest total views
yt.groupby('Channel_title')['views'].sum().sort_values(ascending=False).head(1)
```

```python
# Channel with the highest average views per video
yt.groupby('Channel_title')['views'].mean().sort_values(ascending=False).head(1)
```

```python
# Every channel's total views, ranked
yt.groupby('Channel_title')['views'].sum().sort_values(ascending=False)
```

```python
# Every channel's total likes, ranked
yt.groupby('Channel_title')['likes'].sum().sort_values(ascending=False)
```

```python
# Every channel's average views, ranked
yt.groupby('Channel_title')['views'].mean().sort_values(ascending=False)
```

```python
# Every channel's max views, ranked
yt.groupby('Channel_title')['views'].max().sort_values(ascending=False)
```

```python
# Every channel's video count, ranked
yt['Channel_title'].value_counts()
```

## 7. Engagement Analysis

```python
# Engagement rate = likes / views
yt['engagement_rate'] = yt['likes'] / yt['views']
yt[['views', 'likes', 'engagement_rate']].head()
```

```python
# Top 10 videos by engagement rate
yt.sort_values('engagement_rate', ascending=False).head(10)
```

```python
# Videos with above-average engagement rate
avg_engagement = yt['engagement_rate'].mean()
above_avg = yt[yt['engagement_rate'] > avg_engagement]
print(f'Average engagement rate: {avg_engagement:.4f}')
print(f'Videos above average: {len(above_avg)} of {len(yt)}')
above_avg.head()
```
